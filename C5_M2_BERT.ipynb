```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
# Load pre-trained model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# Two sentences with different meanings of "bank"
sentences = [
    "I need to go to the bank to deposit money.",
    "The river bank was eroding after the flood."
]
```

```python
# Get embeddings for the word "bank" in each context
for sentence in sentences:
    # Tokenize the sentence
    inputs = tokenizer(sentence, return_tensors="pt")
    
    # Find position of "bank" in tokenized input
    bank_tokens = tokenizer.tokenize("bank")
    all_tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
    bank_position = all_tokens.index(bank_tokens[0])
    
    # Get model output
    with torch.no_grad():
        outputs = model(**inputs)
    
    # Get the embedding for "bank" (last hidden state at bank's position)
    bank_embedding = outputs.last_hidden_state[0, bank_position]
    
    print(f"Sentence: {sentence}")
    print(f"Contextual embedding shape: {bank_embedding.shape}")
    print(f"First 5 dimensions: {bank_embedding[:5].numpy()}\n")
```
