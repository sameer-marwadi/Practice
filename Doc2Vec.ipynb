```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
# Sample documents
documents = [
   "Machine learning is revolutionizing the tech industry.",
   "Natural language processing helps computers understand human language.",
   "Deep learning models require large amounts of data to train effectively.",
   "Artificial intelligence systems can now recognize speech and images.",
   "Soccer is a sport and not related to machine learning"
]
```

```python
# Prepare documents for Doc2Vec
tagged_docs = [TaggedDocument(nltk.word_tokenize(doc.lower()), [i])
              for i, doc in enumerate(documents)]
```

```python
# Train Doc2Vec model
model = Doc2Vec(tagged_docs, vector_size=50, window=2, min_count=1, workers=4, epochs=100)
```

```python
# Get vector for a document
doc_vector = model.infer_vector(nltk.word_tokenize(documents[0].lower()))
print(f"Document vector shape: {doc_vector.shape}")
```

```python
# Find similar documents
new_query = "How is artificial intelligence evolving?"
query_vector = model.infer_vector(nltk.word_tokenize(new_query.lower()))
similar_docs = model.dv.most_similar([query_vector], topn=len(documents))
```

```python
print(f"Query: \"{new_query}\"")
print("Similar documents:")
for doc_id, similarity in similar_docs:
   print(f"Document {doc_id}: \"{documents[doc_id]}\"")
   print(f"Similarity: {similarity:.4f}")
```
