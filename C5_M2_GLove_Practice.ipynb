```python
import numpy as np
from gensim.models import KeyedVectors
```

```python
# Load pre-trained GloVe embeddings
# First, download from https://nlp.stanford.edu/projects/glove
glove_file = 'glove.6B.100d.txt'  # Local path to the file
glove_model = {}
```

```python
with open(glove_file, 'r', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.array(values[1:], dtype='float32')
        glove_model[word] = vector

# Create a function to find similar words
def find_similar(model, word, n=5):
    word_vector = model[word]
    similarities = {}
    
    for w, vec in model.items():
        if w != word:
            # Compute cosine similarity
            sim = np.dot(word_vector, vec) / (np.linalg.norm(word_vector) * np.linalg.norm(vec))
            similarities[w] = sim
    
    # Return top n similar words
    return sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:n]

similar_words = find_similar(glove_model, 'computer')
print("Words similar to 'computer':")
for word, similarity in similar_words:
    print(f"  {word}: {similarity:.4f}")
```
